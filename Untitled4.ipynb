### Multivariate Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fuel-econ.csv')
```

```python
df.head()
```

```python
df['trans_type'] = df['trans'].str.split(expand=True)[0]
```

```python
df
```

```python
plt.scatter(data=df, x='displ',y='comb');
```

```python
df
```

```python
df_new = df.loc[numbers]
```

```python
np.random.seed(1)
numbers = np.random.choice(3929, 200, replace=False)
```

```python
df_new.head()
```

```python
sns.regplot(data=df_new, x='displ',y='comb',
           x_jitter=0.04, fit_reg=False)
plt.xlabel('Displacement(l)')
plt.ylabel('Comb Fuel Eff(mpg)')
plt.title('Multivariate Chart');
```

### Shape

```python
cat_shapes = [['Automatic','^'],['Manual','s']]

for cat,shape in cat_shapes:
    sns.regplot(data=df_new[df_new.trans_type==cat],
               x='displ', y='comb',
               fit_reg=False, x_jitter=0.04,
               marker=shape)
plt.legend(['Automatic','Manual']);
```

### Size

```python
df_new.head()
```

```python
sns.regplot(data=df_new, x='displ', y='comb',
           fit_reg=False, x_jitter=0.04,
           scatter_kws= {'s':df_new.co2/4})
plt.xlabel('Displacement(l)')
plt.ylabel('Comb Fuel Eff(mpg)')
plt.title('Multivariate Chart');
```

### Color

### Color Palettes - Qualitative, Sequential, Diverging

```python
sns.color_palette()
```

```python
f = sns.FacetGrid(data=df_new, hue='trans_type', height=4)
f.map(sns.regplot, 'displ','comb',
     x_jitter=0.04, fit_reg=False)
f.add_legend();
```

```python
f = sns.FacetGrid(data=df_new, hue='VClass', height=4, palette='viridis')
f.map(sns.regplot, 'displ','comb',
     x_jitter=0.04, fit_reg=False)
f.add_legend();
```

```python
plt.scatter(data=df_new, x='displ',y='comb', c='co2', cmap='viridis_r')
plt.xlabel('Displacement(l)')
plt.ylabel('Comb Fuel Eff(mpg)')
plt.title('Multivariate Chart')
plt.colorbar();
```
